```python
import random

def valida_int(pergunta, min, max):
    while True:
        try:
            x = int(input(pergunta))
            if min <= x <= max:
                return x
        except ValueError:
            pass
        print(f"Digite um número entre {min} e {max}.")

def nome_jogada(num):
    nomes = {1: 'Pedra', 2: 'Papel', 3: 'Tesoura'}
    return nomes.get(num, 'Inválido')

def vencedor(jogador1, jogador2):
    global v1, v2, empate
    if jogador1 == jogador2:
        empate += 1
        return 'Empate!'
    elif (jogador1 == 1 and jogador2 == 3) or \
         (jogador1 == 2 and jogador2 == 1) or \
         (jogador1 == 3 and jogador2 == 2):
        v1 += 1
        return 'Jogador 1 venceu!'
    else:
        v2 += 1
        return 'Computador venceu!'

# PROGRAMA PRINCIPAL
while True:
    print('\nJOKENPO')
    print('1 - Pedra')
    print('2 - Papel')
    print('3 - Tesoura')
    print('0 - Sair do jogo')

    jogadas = []
    v1 = 0
    v2 = 0
    empate = 0

    while True:
        j1 = valida_int('\nEscolha sua ação (0 para sair): ', 0, 3)
        if j1 == 0:
            break

        j2 = random.randint(1, 3)

        resultado_rodada = vencedor(j1, j2)
        jogadas.append((j1, j2, resultado_rodada))

        print(f'Você jogou: {nome_jogada(j1)} | Computador jogou: {nome_jogada(j2)}')
        print(f'Resultado: {resultado_rodada}')
        print(f'Placar atual - Jogador 1: {v1} | Computador: {v2} | Empates: {empate}')

    print('\n=== RESULTADO FINAL ===')
    for index, jogada in enumerate(jogadas, 1):
        print(f'Rodada {index}: Jogador 1 → {nome_jogada(jogada[0])} x Computador → {nome_jogada(jogada[1])} | {jogada[2]}')

    print(f'\nTotal de vitórias do Jogador 1: {v1}')
    print(f'Total de vitórias do Computador: {v2}')
    print(f'Total de empates: {empate}')

    jogar_novamente = input('\nQuer jogar novamente? (s/n): ').strip().lower()
    if jogar_novamente != 's':
        print('Obrigado por jogar! Até a próxima!')
        break
```
